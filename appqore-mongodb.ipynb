```python
import json
import glob
import os
from pymongo import MongoClient

# 1. Setup connection
uri = "mongodb://127.0.0.1:27017/"
client = MongoClient(uri)
db = client["saas_data"]
collection = db["provider_export"]

# 2. Update this line to target the specific folder
# Using os.path.join is a best practice for cross-platform compatibility
folder_path = "appqore-policies"
json_files = glob.glob(os.path.join(folder_path, "*.json"))

for file_path in json_files:
    print(f"Processing {file_path}...")
    
    # 3. Read file line-by-line
    with open(file_path, 'r') as f:
        batch = []
        for line_number, line in enumerate(f, 1):
            line = line.strip()
            if not line:
                continue  # Skip empty lines
                
            try:
                # Parse the individual line
                doc = json.loads(line)
                batch.append(doc)
                
                # Optional: Insert in chunks of 500 for better performance
                if len(batch) >= 500:
                    collection.insert_many(batch)
                    batch = []
                    
            except json.JSONDecodeError:
                print(f"Error: Invalid JSON on line {line_number} in {file_path}")

        # 4. Insert any remaining documents in the final batch
        if batch:
            collection.insert_many(batch)

print("\nImport complete!")
print(f"Total documents in collection: {collection.count_documents({})}")
```
