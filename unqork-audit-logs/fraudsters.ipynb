```python
import pandas as pd
```

```python
df_user_ipaddress = pd.read_csv('2025-december-output.csv')
```

```python
df_user_ipaddress.columns
```

```python
ip_counts = df_user_ipaddress.groupby(['userid', 'ipaddress'])['submissionid'].nunique().reset_index()
```

```python
ip_counts.columns = ['Agent', 'IP_Address', 'App_Count']
```

```python
THRESHOLD = 3
suspicious_agents = ip_counts[ip_counts['App_Count'] >= THRESHOLD].sort_values(by='App_Count', ascending=False)
```

```python
suspicious_agents
```

```python
suspicious_agents.to_csv('suspicious_agents.csv', index=False)
```
