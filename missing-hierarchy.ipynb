```python
import pandas as pd
df_appqore_policies = pd.read_csv('appqore-policies-2251104.csv')
df_missing_hierarchy = pd.read_csv('missing-hierarchy-20251104.csv')
```

```python
df_appqore_policies
```

```python
policies_df = pd.merge(df_appqore_policies.rename(columns={'policyNumber': 'policyNumber_left'}), df_missing_hierarchy.rename(columns={'policyNumber': 'policyNumber_right'}), left_on='policyNumber_left', right_on='policyNumber_right', how='outer')
```

```python
policies_df.to_csv('appqore-missing-hierarchy-two-months-20251104.csv', index=False)
```
